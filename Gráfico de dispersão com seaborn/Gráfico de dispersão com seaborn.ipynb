# _Gráfico de dispersão com seaborn_

```python
# Bibliotecas
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# Carregamento dos dados
df = pd.read_csv('co2.csv')
df.head()
```

```python
# Gráfico de dispersão utilizando 'conc' e 'uptake', agrupado pelo 'type'
sns.scatterplot(df['conc'], df['uptake'], hue=df['Type']);
# plt.tight_layout();
```

```python
# Seleção específica dos registros na base de dados
q = df.loc[df['Type']=='Quebec']
m = df.loc[df['Type']=='Mississippi']
```

```python
# Gráfico sobre cada região
plt.figure(figsize=(10, 6))
plt.subplot(1,2,1)
sns.scatterplot(q.conc, q.uptake).set_title('Quebec')
plt.subplot(1,2,2)
sns.scatterplot(m.conc, m.uptake).set_title('Mississippi')
plt.tight_layout();
```

```python
# refrigerado e não-refrigerado
re = df.loc[df['Treatment']=='chilled']
nr = df.loc[df['Treatment']=='nonchilled']
```

```python
# Gráfico sobre refrigeração
plt.figure(figsize=(10, 6))
plt.subplot(1,2,1)
sns.scatterplot(re.conc, re.uptake).set_title('chilled')
plt.subplot(1,2,2)
sns.scatterplot(nr.conc, nr.uptake).set_title('nonchilled')
plt.tight_layout();
```

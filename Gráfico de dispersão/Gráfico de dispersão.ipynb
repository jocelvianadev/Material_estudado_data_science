# _Gráfico de dispersão_
_É excelente para comparar duas variáveis numéricas_

```python
# Bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Base de dados
df = pd.read_csv('trees.csv')
df.head()
```

```python
# Gráfico de dispersão considerando o volume e a dispersão
plt.scatter(x=df['Girth'], y=df['Volume'], color='Red', facecolors = 'none', marker='*')
plt.title('Árvores')
plt.xlabel('Circunferência')
plt.ylabel('Volume')
plt.tight_layout();
```

```python
# Gráfico de linhas considerando o volume e o atributo 'Girth' sem marker
plt.plot(df['Girth'], df['Volume'])
plt.title('Árvores')
plt.xlabel('Circunferência')
plt.ylabel('Volume')
plt.tight_layout();
```

#### _Usando o searborn_

```python
# Gráfico de dispersão com 'afastamento' dos dados (jitter)
# fit_reg = linha de tendência
# jitter = tremulação dos dados
sns.regplot(x=df['Girth'], y=df['Volume'], data = df, x_jitter=0.3, fit_reg = True);
```

### _Gráfico de dispersão com legenda_

```python
# Outra base de dados
df2 = pd.read_csv('co2.csv')
df2.head()
```

```python
# Criando duas variáveis para cada atributo
x = df2['uptake']
y = df2['conc']
```

```python
df2.shape
```

```python
# Retorna os valores únicos do atributo Treatment
unicos = list(set(df2['Treatment']))
unicos
```

```python
# Percorre cada tipo de tratamento e cria um gráfico de dispersão
for i in range(len(unicos)):
    indice = df2['Treatment'] == unicos[i]
    plt.scatter(y[indice],x[indice], label = unicos[i])
plt.legend(loc = 'lower right');
```

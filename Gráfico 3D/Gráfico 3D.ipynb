# _Gráfico 3D_

```python
# Bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d # Para o gráfico 3D
```

```python
# Base de dados
df = pd.read_csv('orchard.csv')
df.head()
```

```python
# Gráfico de dispersão
figura = plt.figure()
eixo = figura.add_subplot(1,1,1, projection = '3d')
eixo.scatter(df['decrease'], df['rowpos'], df['colpos'])
eixo.set_xlabel('decrease')
eixo.set_ylabel('rowpos')
eixo.set_zlabel('colpos')
plt.tight_layout();
```
